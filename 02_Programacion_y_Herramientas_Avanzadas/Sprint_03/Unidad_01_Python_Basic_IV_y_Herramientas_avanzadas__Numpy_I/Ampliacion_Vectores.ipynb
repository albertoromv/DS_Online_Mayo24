3. Aplicamos la definicin de **distancia** entre vectores para relacionar los alumnos entre s.


```python
import math

def distancia_2d(vec1, vec2, precision = 2): # Esperamos vectores de 2 dimensiones, es decir "listas" con longitud 2
    suma = 0
    for indice_componente in range(2): # 2 por ser la cardinalidad (nmero de componentes del vector)
        suma += (vec1[indice_componente] - vec2[indice_componente])**2
    distancia = math.sqrt(suma)
    distancia = round(distancia, precision)
    return distancia
```

### El producto escalar de vectores  

[al indice](#Contenidos)  



El [producto escalar](https://es.wikipedia.org/wiki/Producto_escalar) de dos vectores se define como la suma de los productos de sus elementos, suele representarse matemticamente como < x, y > o x'y, donde x e y son dos vectores.

$$< x, y > := \sum_{i=1}^n x_i y_i$$


Con un ejemplo sobre nuestra clase: 

$$< Valeria , Alejandro > := \sum_{i=0}^1 Valeria_i * Alejandro_i$$

Como:

$Valeria = [-5,4]$  
$Alejandro = [3, 5]$

Sustituyendo:

$$< Valeria , Alejandro > := (-5 * 3) + (4 * 5) = -15 + 20 = 5$$

El producto escalar de los vectores que caracterizan a Valeria y a Alejandro es 5. Fijate que el resultado no es otro vector sino un nmero, por eso se denomina *producto escalar*

En **terminologa de Algebra**, un nmero (cualquier nmero) es un **"escalar"** [nada que ver con nuestra clase de escalar con robots, no nos hagamos un lo], y una secuencia de escalares es un **"vector"**. Por ejemplo:

* $12$, $-21$ y $\sqrt2$ cada uno son *escalares* 
* $[12,-21]$ y $[\sqrt3, -0.12, 1/5]$ son *vectores*

### Normas, magnitudes y longitudes  

[al indice](#Contenidos)  



En realidad son conceptos anlogos en el caso de vectores (y en concreto de vectores con menos de cuatro dimensiones o componentes). Recuerda que en la sesin anterior definimos los vectores como: una **entidad matemtica** que tiene tanto magnitud (o longitud) como direccin.

La norma es otra forma de denominar a esa magnitud o esa longitud, es decir lo "largo" que es el vector (concepto que podeos observar perfectamente cuando el vector tiene 2 dimensiones como es el caso de las caractersticas de los alumnas y alumnos de la profesora Aidone).



Lo formal, es decir que todo producto escalar induce una [norma](https://es.wikipedia.org/wiki/Norma_vectorial) sobre el espacio en el que est definido, pero no hace falta que te quedes con esto. La frmula de la norma de un vector es:

$$\| x \| := \sqrt{< x, x>} := \left( \sum_{i=1}^n x_i^2 \right)^{1/2}$$




Es decir la raiz cuadrada del producto escalar del vector por s mismo... (no te rompas la cabeza con ello, quedate con que es as y es la forma de obtener la longitud del vector)



Para Lucia ( $[ x_1 = 2, x_2 = -3]$):

$||Lucia|| = \sqrt{<Lucia,Lucia>} = \sqrt{(x_1 * x_1)+(x_2 * x_2)} = \sqrt{( 2 * 2 )+(-3 * -3)} = \sqrt{4+9} = \sqrt{13} \approx 3.6$

### Angulo entre dos vectores  

[al indice](#Contenidos)  



Ya casi estamos, nos queda saber cul es la relacin entre el producto escalar entre dos vectores y el ngulo o una funcin del ngulo que forman esos dos vectores.

Esa relacin viene dada por esta funcin (tampoco hace falta que la memorices, slo que sepas de su existencia):

$<vec1, vec2> = vec1 \cdot vec2 = ||vec1|| * ||vec2|| * cos(\theta)$

Donde $vec_1$ y $vec_2$ son los dos vectores de los que queremos saber el ngulo o una funcin del ngulo y $\theta$ es el ngulo que forman esos dos vectores.

Despejando:

$cos(\theta)= \frac{vec1 \cdot vec2}{||vec1|| * ||vec2||}$

Es decir que conociendo las normas y el producto escalar podemos obtener el coseno del angulo que forman. Sin que tengas que recordar la trigonometra de la ESO, el bachillerato, el EGB, o lo que sea, slo debes recordar que el coseno de un ngulo tiene valores entre -1 y 1 y que es mayor a medida que el ngulo es ms pequeo y menor a medida que el ngulo es mayor.

En concreto:

$cos(0) = 1$ -> Vectores superpuestos   
$cos(90) = 0$ -> Vectores perpendiculares  
$cos(180) = -1$ -> Vectores apuntando en sentidos contrarios

Grficamente:



En definitiva que a medida que el coseno se hace ms pequeo menor es la similitud y viceversa.

Apliqumoslo sobre un ejemplo de nuestro problema, obtengamos el coseno del ngulo ($\theta$) entre Lucia y Gabriela:

Partiendo de:  

Lucia = $[Vuelo = 2, Comida = -3]$  
Gabriela = $[Vuelo = 5, Comida = -2]$

Aplicamos:  

$cos(\theta) = \frac{Lucia \cdot Gabriela}{||Lucia||*||Gabriela||}$
  

Por partes: 
* $||Lucia|| = \sqrt{(2 * 2) + (-3 * -3)} \approx 3.6$
* $||Gabriela|| = \sqrt{(5 * 5) + (-2 * -2)} \approx 5.4$
* $Lucia \cdot Gabriela = (2 * 5) + (-3 * -2) = 16$

$cos(\theta) = \frac{16}{3.6 * 5.4} \approx 0.82$

El resultado es de 0.82 bastante cercano a 1, cmo esperbamos porque el ngulo que separa a Lucia y a Gabriela parece pequeo. Puedes comprobar a mano que los cosenos de los ngulos de Lucia con Rodrigo y Mateo son menores.
