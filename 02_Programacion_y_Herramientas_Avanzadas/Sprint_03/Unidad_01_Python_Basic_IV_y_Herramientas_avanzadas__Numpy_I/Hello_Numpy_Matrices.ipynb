```python
matriz_asignaciones = [[2, 1, 0.5], 
                       [1, 2, 0.5], 
                       [2, 1, 0.5], 
                       [1, 0.5, 2], 
                       [0.5, 2, 1], 
                       [2, 1, 0.5], 
                       [0.5, 1, 2], 
                       [1, 2, 0.5], 
                       [1, 0.5, 2], 
                       [0.5, 1, 2]
]
```

### Doblar las horas de las clases no prioritarias, aumentar una hora y media a la clase extraescolar prioritaria  
[al indice](#Contenidos)  


Recuerda que habíamos llegado a que la solución con matrices era:

$$ NuevaAsignacion =  2 * \begin{bmatrix}2&1&0.5\\1&2&0.5\\2&1&0.5\\1&0.5&2\\0.5&2&1\\2&1&0.5\\0.5&1&2\\1&2&0.5\\1&0.5&2\\0.5&1&2\end{bmatrix} - \begin{bmatrix}0.5&0&0\\0&0.5&0\\0.5&0&0\\0&0&0.5\\0&0.5&0\\0.5&0&0\\0&0&0.5\\0&0.5&0\\0&0&0.5\\0&0&0.5\end{bmatrix}$$

```python
# Creamos la matriz resta
matriz_resta = []

for vector in matriz_asignaciones:
    new_vector = [0,0,0]
    new_vector[vector.index(2)] = 0.5
    matriz_resta.append(new_vector)

matriz_resta_np = np.array(matriz_resta)
```
